```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt

freq = 12
rate = 0.12
rate_house = 0.05
years = 5
pv = 120000

rate /= freq  # konwersja stopy do okresu miesięcznego
nper = years * freq  # liczba wszystkich okresów
periods = np.arange(1,nper+1,dtype=int)
```

```python
price_in_5 = round(pv *(1 + rate_house) ** years, 2)
price_in_5
```

```python
month_payment = round(npf.pmt(rate, nper, 0, -price_in_5),2)
month_payment
```

```python
miesiace_os = np.arange(1, nper + 1)
ceny_mieszkania = pv * (1 + rate_house) ** (miesiace_os / 12)
wartosc_lokaty = npf.fv(rate, miesiace_os, -month_payment, 0)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(miesiace_os, ceny_mieszkania, label="Cena mieszkania", color="blue")
plt.plot(miesiace_os, wartosc_lokaty, label="Wartość lokaty", color="green", linestyle="--")
plt.xlabel("Miesiące")
plt.ylabel("Wartość [PLN]")
plt.title("Cena mieszkania vs Wartość lokaty")
plt.legend()
plt.grid()
plt.show()
```

```python
print(f"Orientacyjna cena mieszkania za 5 lat: {price_in_5:.2f} PLN")
print(f"Miesięczna kwota wpłat na lokatę: {month_payment:.2f} PLN")
```
